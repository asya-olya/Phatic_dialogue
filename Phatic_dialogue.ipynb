```
import random
import re

# dictionary with key words
dictionary = {"film": ["What's your favorite film by the way?","Have you watched The Gentlemen?"],
              "like": ["What else do you like to watch?"],
              "hate": ["What else do you hate watching?"],
              "watched": ["How often do you watch TV series?"],
              "have": ["Yes, I have.","No, I haven`t."],
              "are": ["Yes, I am.","No."],
              "were": ["Yes, I was.","No, I wasn`t."],
              "what": ["I don`t know."],
              "how": ["Actualy, I don`t remember... Somehow."],
              "when": ["Years ago.", "Yesterday.", "Today.", "Never."],
              "where": ["Somewhere.","At home."],
              "why": ["It`s hard to explain. Maybe another time."],
              "ok": ["So what was the last thing you watched?"]
              }

# random greetings
r_greetings = ["Hello!",
             "Hi!",
             "Hey, sweetie!",
             "Good day!",
             "Greetings, darling!",
             "You look AWESOME! What`s new?",
             "How do you do?",
             "How's it going?"]

# random answers for questions
r_answers_for_questions = ["I need to think about this.",
                           "It`s not like I don`t want to, but... Let`s discuss it later.",
                           "Let's not talk about me."]

# random answers
r_answers = ["Very interesting.",
           "I don`t understand...",
           "Let`s change the subject."]

# keyword search
def check(reply):
  result = ""
  for key in dictionary:
     word = dictionary[key]
     if re.search(key, reply.lower()):
       result = random.choice(word)
  
  if result == "":
    result = random.choice(r_answers)

  return result

# generating part of reply using part of a question
def generate_reply(starts, reply):
    
    reply = re.sub(r"(?i)^([^\.\?!]*).*", r"\1?", reply)
    bot_reply = random.choice(starts) + " " + reply
    return bot_reply

# reply for a question with "Can you"
def can_you_reply(reply):
    
    # starts for dynamically generated replies
    starts = ["I am not sure that I can",
              "Sometines I want to",
              "Do you want me to be able to"]

    reply = re.sub(r"(?i)^(can|do) you ", r"", reply)

    # choose random between dynamic and random answer
    return random.choice([random.choice(r_answers_for_questions),
                          generate_reply(starts, reply)])

# reply for a question with "Can I"
def can_I_reply(reply):

    # starts for dynamically generated replies
    starts = ["Why do you want to",
              "Why are you asking me to",
              "Do you really want to"]

    reply = re.sub(r"(?i)^(can|may) I ", r"", reply)

    # choose random between dynamic and random answer 
    return random.choice(["Yes, of course!", "Maybe.", "No way!", "It makes sense.",
                          generate_reply(starts, reply)])

############################################################################################################################################

#start
print("Today I want to talk about films and series with you.")
print("If you want me to stop talking to you, just tell me about it using the word filmed in your sentence. Let`s start!")
print(random.choice(r_greetings))
#

#
reply = input()
while not re.search(r"\bfilmed\b", reply.lower()):  

      if re.search(r"(?i)^(can|do) you ", reply):
        print(can_you_reply(reply))
      else:
         if re.search(r"(?i)^(can|may) I ", reply):
           print(can_I_reply(reply)) 
         else:
           print(check(reply))

      reply = input()
```
